```python
import json
import numpy as np
import matplotlib.pyplot as plt
import japanize-matplotlib
import seaborn as sns

# 1. JSONファイルの読み込み
# ※ 先ほどのJSONデータを 'result.json' という名前で同じフォルダに保存してください
file_path = '/home/bekuta/RB2023/shikama/api2img-output/result/img_gray/20260319-014230/train_log.json'

try:
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
except FileNotFoundError:
    print(f"エラー: {file_path} が見つかりません。JSONファイルを保存してください。")
    exit()

# 2. 各Runの混同行列を取得して合算する
# 初期化 (2x2のゼロ行列)
cumulative_cm = np.zeros((2, 2), dtype=int)

# 'runs' の中から 'confusion_matrix' を取り出して足し合わせる
for run_data in data['runs']:
    cm = np.array(run_data['confusion_matrix'])
    cumulative_cm += cm

print("【集計完了】累積混同行列:")
print(cumulative_cm)

# 3. ヒートマップとして描画 (可視化)
# ラベルの設定 (JSONの 'label_order' に基づく: 0=Benign, 1=Malware)
classes = ['Benign', 'Malware']

plt.figure(figsize=(8, 6))
sns.set_theme(style='whitegrid', font_scale=1.2)

# seabornのヒートマップ作成
ax = sns.heatmap(
    cumulative_cm, 
    annot=True,      # 数値を表示
    fmt='d',         # 整数フォーマット
    cmap='Blues',    # 青系のカラーマップ
    cbar=False,      # カラーバーを非表示（スッキリさせるため）
    xticklabels=classes, 
    yticklabels=classes,
    annot_kws={"size": 16, "weight": "bold"} # 文字のサイズと太さ
)

# タイトルと軸ラベルの設定
plt.title('Cumulative Confusion Matrix (Total 5 Runs)', pad=20, fontsize=16, fontweight='bold')
plt.xlabel('Predicted Label', labelpad=10, fontsize=14)
plt.ylabel('True Label', labelpad=10, fontsize=14)

# レイアウトを整えて表示
plt.tight_layout()
plt.show()
```
